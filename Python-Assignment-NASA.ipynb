```python
# Task A: 3D Plot of Impedance Measurements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Generate Random EIS Data
rng = np.random.default_rng(123)  
num_cycles = 100
cycle_index = np.arange(1, num_cycles + 1)
real_z = rng.uniform(0, 120, num_cycles)
imag_z = rng.uniform(0, 40, num_cycles)

# 2. Create a DataFrame
eis_df = pd.DataFrame({
    "Cycle_Index": cycle_index,
    "Real_Impedance_kOhms": real_z,
    "Imag_Impedance_kOhms": imag_z
}).sort_values("Cycle_Index")

# 3. Set up a 3D Figure
fig = plt.figure(figsize=(10, 7))
ax3d = fig.add_subplot(111, projection='3d')

# 4. Plot the Data in 3D
scatter3d = ax3d.scatter(
    eis_df["Real_Impedance_kOhms"],
    eis_df["Imag_Impedance_kOhms"],
    eis_df["Cycle_Index"],
    c=eis_df["Cycle_Index"],
    cmap="viridis",   
    s=60,
    alpha=0.8
)

# 5. Label the Axes and Add a Title
ax3d.set_xlabel("R(Z) (kΩ)")
ax3d.set_ylabel("Im(Z) (kΩ)")
ax3d.set_zlabel("Cycle Index")
ax3d.set_title("3D EIS Visualization: Impedance vs. Cycle", pad=20)

# 6. Adjust the View and Add a Color Bar
ax3d.view_init(elev=20, azim=140)  
colorbar = fig.colorbar(scatter3d, ax=ax3d, shrink=0.7)
colorbar.set_label("Cycle Index")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ----- Task B: Incremental Capacity Analysis (ICA) -----
# Generate a dataset for charge/discharge cycle analysis
np.random.seed(42)
voltage = np.linspace(3.1, 3.6, 100)  # Simulating voltage range
charge_capacity = np.cumsum(np.random.uniform(0.01, 0.1, len(voltage)))  # Simulated charge capacity
discharge_capacity = charge_capacity[::-1] + np.random.normal(0, 0.02, len(voltage))  # Simulated discharge capacity

# Calculate incremental capacity (dQ/dV)
incremental_capacity_charge = np.gradient(charge_capacity, voltage)
incremental_capacity_discharge = np.gradient(discharge_capacity, voltage)

# Create DataFrame for visualization
data_cycles = pd.DataFrame({
    "Voltage (V)": voltage,
    "Charge Capacity (Ah)": charge_capacity,
    "Discharge Capacity (Ah)": discharge_capacity,
    "dQ/dV Charge (Ah/V)": incremental_capacity_charge,
    "dQ/dV Discharge (Ah/V)": incremental_capacity_discharge
})

# 2D Plot: Incremental Capacity Analysis
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data_cycles["Voltage (V)"], data_cycles["dQ/dV Charge (Ah/V)"], label="IC - Charge", color="blue")
ax.plot(data_cycles["Voltage (V)"], data_cycles["dQ/dV Discharge (Ah/V)"], label="IC - Discharge", color="red")
ax.set_xlabel("Voltage (V)")
ax.set_ylabel("Incremental Capacity (Ah/V)")
ax.set_title("Incremental Capacity Analysis (dQ/dV vs Voltage)")
ax.legend()
plt.show()

# Simulating cycle count and peaks for 3D visualization
cycle_counts = np.arange(1, 101)  # Simulating 100 cycles
peak_voltages = np.random.choice(voltage, len(cycle_counts))
peak_dq_dv = np.random.uniform(0.5, 1.5, len(cycle_counts))

# Create DataFrame for 3D plot
data_3d = pd.DataFrame({
    "Cycle Count": cycle_counts,
    "Peak Voltage (V)": peak_voltages,
    "Peak dQ/dV (Ah/V)": peak_dq_dv
})

# 3D Plot: Visualize peak changes with aging
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_3d["Peak Voltage (V)"], data_3d["Peak dQ/dV (Ah/V)"], data_3d["Cycle Count"], 
           c='green', label='Peak Analysis')
ax.set_xlabel("Peak Voltage (V)")
ax.set_ylabel("Peak dQ/dV (Ah/V)")
ax.set_zlabel("Cycle Count")
ax.set_title("3D Visualization of Incremental Capacity Peaks")
plt.legend()
plt.show()

voltage_charge = np.linspace(3.1, 3.45, 100)
dq_dv_charge = np.sin(10 * (voltage_charge - 3.25)) * np.exp(-(voltage_charge - 3.25)**2 / 0.01)

voltage_discharge = np.linspace(3.1, 3.45, 100)
dq_dv_discharge = -np.sin(10 * (voltage_discharge - 3.25)) * np.exp(-(voltage_discharge - 3.25)**2 / 0.01)

plt.figure(figsize=(10, 6))
plt.plot(voltage_charge, dq_dv_charge, label="IC - Charge", color='blue')
plt.plot(voltage_discharge, dq_dv_discharge, label="IC - Discharge", color='red')
plt.xlabel("Voltage (V)")
plt.ylabel("Incremental Capacity (dQ/dV)")
plt.title("Improved Incremental Capacity Analysis (Charge and Discharge)")
plt.legend()
plt.grid()
plt.show()

# 3D Plot of ICA Peaks with Aging
cycle_count = np.arange(1, 101)  # 100 cycles
ica_peaks_charge = [np.max(dq_dv_charge) - (i * 0.01) for i in range(100)]
ica_peaks_discharge = [np.min(dq_dv_discharge) + (i * 0.01) for i in range(100)]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(cycle_count, ica_peaks_charge, zs=3.35, zdir='z', label="Charge ICA Peaks", color='blue')
ax.plot(cycle_count, ica_peaks_discharge, zs=3.15, zdir='z', label="Discharge ICA Peaks", color='red')
ax.set_xlabel("Cycle Count")
ax.set_ylabel("ICA Peaks (dQ/dV)")
ax.set_zlabel("Voltage (V)")
ax.set_title("3D Plot of ICA Peaks with Aging")
plt.legend()
plt.show()
```

```python
# Task C: Predicting Current Capacity with Examples

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # (unused in ML portion, but retained if needed later)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ----- Generate Synthetic Data for EIS and Battery Capacity Prediction -----
np.random.seed(42)
n_samples = 500
# Simulate impedance features (in kΩ)
impedance_real = np.random.uniform(0, 120, n_samples)
impedance_imag = np.random.uniform(0, 40, n_samples)

capacity_actual = 50 - 0.2 * impedance_real - 0.1 * impedance_imag + np.random.normal(0, 1, n_samples)

# Build DataFrame for ML
df_ml = pd.DataFrame({
    "Real_Impedance": impedance_real,
    "Imag_Impedance": impedance_imag,
    "Capacity": capacity_actual
})

# ----- Split Data into Training and Testing Sets -----
features = df_ml[["Real_Impedance", "Imag_Impedance"]]
target = df_ml["Capacity"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

param_options = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

gbr_model = GradientBoostingRegressor(random_state=42)
grid_cv = GridSearchCV(estimator=gbr_model, param_grid=param_options, cv=5, 
                       scoring='neg_mean_squared_error', verbose=1)
grid_cv.fit(X_train, y_train)

optimal_model = grid_cv.best_estimator_
print("Optimal Parameters:", grid_cv.best_params_)

# ----- Predict and Evaluate the Model -----
predicted_capacity = optimal_model.predict(X_test)

mse_val = mean_squared_error(y_test, predicted_capacity)
mae_val = mean_absolute_error(y_test, predicted_capacity)
r2_val = r2_score(y_test, predicted_capacity)

print(f"MSE: {mse_val:.4f}")
print(f"MAE: {mae_val:.4f}")
print(f"R2 Score: {r2_val:.4f}")

# ----- Display Sample Predictions -----
sample_results = []
n_examples = min(5, len(X_test))
for i in range(n_examples):
    sample_feat = X_test.iloc[i:i+1]
    actual_val = y_test.iloc[i]
    pred_val = optimal_model.predict(sample_feat)[0]
    sample_results.append({
        "Index": i,
        "Features": sample_feat.iloc[0].to_dict(),
        "Actual Capacity": actual_val,
        "Predicted Capacity": pred_val,
        "Error": actual_val - pred_val
    })

results_df = pd.DataFrame(sample_results)
print(results_df)

# ----- Plotting: Actual vs Predicted Capacity -----
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predicted_capacity, alpha=0.7, label="Predicted")
plt.plot([target.min(), target.max()], [target.min(), target.max()], 'r--', lw=2, label="Ideal Fit")
plt.xlabel("Actual Capacity")
plt.ylabel("Predicted Capacity")
plt.title("Actual vs. Predicted Battery Capacity")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ----- Print Additional Metrics -----
print("Full Test Set Evaluation:")
print(f"RMSE: {np.sqrt(mse_val):.4f}")
print(f"MAE: {mae_val:.4f}")
```
